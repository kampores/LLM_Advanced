# Session 17: Continuous Pretraining with Qwen2.5-1.5B (FFT)

## 🎯 Continuous Pretraining 개요

**Continuous Pretraining(지속 사전학습)**은 이미 사전 학습된 LLM에 도메인 특화 텍스트를 추가로 학습시키는 방법입니다.

### 왜 Continuous Pretraining이 필요한가?

- 🔹 사전학습 모델은 일반적인 지식은 풍부하지만, 특정 도메인 지식이 부족할 수 있습니다
- 🔹 의료, 법률, 금융 등 전문 분야의 용어와 지식을 모델에 주입할 수 있습니다
- 🔹 Instruction Tuning과 달리, **비구조화된 텍스트**(plain text)를 그대로 학습합니다
- 🔹 Next Token Prediction 방식으로 도메인 언어 패턴을 학습합니다

### 왜 FFT(Full Fine-Tuning)인가?

- 🔹 CPT는 모델의 **기본 지식 자체를 변경**하는 작업입니다
- 🔹 LoRA처럼 일부 파라미터만 건드리면 새 지식이 깊이 반영되기 어렵습니다
- 🔹 FFT로 전체 파라미터를 업데이트해야 도메인 지식이 제대로 주입됩니다
- 🔹 1.5B 모델은 float16 기준 ~3GB로, RTX 4060(8GB)에서 FFT가 가능합니다

### 학습 목표

- ✅ Continuous Pretraining의 개념과 활용 사례 이해
- ✅ 도메인 텍스트 데이터 전처리 (토큰화 + 청크 분할)
- ✅ FFT를 활용한 Continuous Pretraining 수행
- ✅ 학습 전후 도메인 지식 변화 비교

### 실습 환경

| 항목 | 설정 |
|------|------|
| 모델 | Qwen2.5-1.5B (float16) |
| 방법 | FFT (Full Fine-Tuning) |
| GPU | RTX 4060 (8GB VRAM) |
| 데이터 | 도메인 텍스트 (AI/ML 관련) |

## 1️⃣ 환경 설정 및 모델 로드 (Qwen2.5-1.5B, float16)

````python
# 필수 라이브러리 임포트
import torch
import gc
import os
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling
)
from datasets import Dataset
import warnings
warnings.filterwarnings('ignore')

print("✅ 라이브러리 임포트 완료")
print(f"PyTorch 버전: {torch.__version__}")
print(f"CUDA 사용 가능: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
````

````python
# GPU 메모리 모니터링 함수
def print_gpu_memory(tag=""):
    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**3
        total = torch.cuda.get_device_properties(0).total_memory / 1024**3
        print(f"[{tag}] GPU: {allocated:.1f}GB / {total:.1f}GB")

print_gpu_memory("시작")
````

````python
# 모델 설정
MODEL_NAME = "Qwen/Qwen2.5-1.5B"  # Instruct 버전이 아닌 base 모델 사용
OUTPUT_DIR = "./output/continuous_pretraining"

print(f"📌 모델: {MODEL_NAME}")
print(f"📌 학습 방식: FFT (Full Fine-Tuning)")
print(f"📌 정밀도: float16")
````

````python
# 토크나이저 로드
print("⏳ 토크나이저 로딩 중...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)

# pad_token 설정
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print(f"✅ 토크나이저 로드 완료")
print(f"📌 Vocab 크기: {tokenizer.vocab_size:,}")
print(f"📌 EOS 토큰: {tokenizer.eos_token}")
print(f"📌 PAD 토큰: {tokenizer.pad_token}")
````

````python
# 모델 로드 (float16)
print("⏳ 모델 로딩 중... (float16)")
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

# gradient checkpointing 활성화 (메모리 절약)
model.gradient_checkpointing_enable(gradient_checkpointing_kwargs={"use_reentrant": False})

print("✅ 모델 로드 완료")
print(f"📌 모델 파라미터 수: {model.num_parameters():,}")
print(f"📌 전체 파라미터가 학습 대상 (FFT)")
print_gpu_memory("모델 로드 후")
````

## 2️⃣ 도메인 텍스트 데이터 준비

Continuous Pretraining에서는 **비구조화된 텍스트**(plain text)를 사용합니다.

- 📄 Instruction-Output 쌍이 아닌 일반 텍스트
- 📄 도메인 특화 문서, 논문, 매뉴얼 등
- 📄 모델이 Next Token Prediction으로 도메인 언어를 학습

````python
# 도메인 텍스트 데이터 로드
data_path = "../data/samples/domain_text_sample.txt"

print("⏳ 도메인 텍스트 데이터 로딩 중...")
with open(data_path, "r", encoding="utf-8") as f:
    domain_text = f.read()

print(f"✅ 데이터 로드 완료")
print(f"📌 전체 텍스트 길이: {len(domain_text):,} 글자")
print(f"📌 줄 수: {len(domain_text.splitlines())}줄")
print(f"\n--- 데이터 미리보기 (처음 500자) ---")
print(domain_text[:500])
````

````python
# 텍스트를 문단 단위로 분할
paragraphs = [p.strip() for p in domain_text.split("\n\n") if p.strip()]

print(f"✅ 문단 분할 완료")
print(f"📌 총 문단 수: {len(paragraphs)}")
print(f"\n--- 문단별 길이 ---")
for i, p in enumerate(paragraphs[:5]):
    print(f"  문단 {i+1}: {len(p)}자 - {p[:60]}...")
````

## 3️⃣ 텍스트 데이터를 토큰화 및 청크 분할

Continuous Pretraining에서는 텍스트를 일정 길이의 **청크(chunk)**로 분할합니다.

- 🔹 max_seq_length 이내로 텍스트를 자릅니다
- 🔹 각 청크가 하나의 학습 샘플이 됩니다
- 🔹 Next Token Prediction: input_ids와 labels가 동일합니다

````python
MAX_SEQ_LENGTH = 1024  # RTX 4060 안전 설정

# 전체 텍스트를 하나로 합치고 토큰화
full_text = "\n\n".join(paragraphs)
print(f"⏳ 텍스트 토큰화 중...")

tokenized = tokenizer(
    full_text,
    return_tensors=None,
    add_special_tokens=False
)

all_input_ids = tokenized["input_ids"]
print(f"✅ 토큰화 완료")
print(f"📌 전체 토큰 수: {len(all_input_ids):,}")
print(f"📌 max_seq_length: {MAX_SEQ_LENGTH}")
````

````python
# 토큰을 max_seq_length 크기의 청크로 분할
chunks = []
for i in range(0, len(all_input_ids), MAX_SEQ_LENGTH):
    chunk = all_input_ids[i:i + MAX_SEQ_LENGTH]
    if len(chunk) >= 64:  # 너무 짧은 청크는 제외
        chunks.append(chunk)

print(f"✅ 청크 분할 완료")
print(f"📌 총 청크 수: {len(chunks)}")
print(f"📌 각 청크 길이: {[len(c) for c in chunks]}")
````

````python
# HuggingFace Dataset 형태로 변환
dataset_dict = {
    "input_ids": chunks,
    "attention_mask": [[1] * len(c) for c in chunks],
    "labels": chunks  # Causal LM: labels = input_ids
}

dataset = Dataset.from_dict(dataset_dict)

print(f"✅ Dataset 생성 완료")
print(f"📌 학습 샘플 수: {len(dataset)}")
print(f"📌 Dataset 컬럼: {dataset.column_names}")
print(f"\n--- 첫 번째 샘플 디코딩 (처음 200자) ---")
print(tokenizer.decode(dataset[0]["input_ids"][:100]))
````

## 4️⃣ FFT 설정

Full Fine-Tuning은 모델의 **전체 파라미터를 업데이트**합니다.

| 항목 | 설정 |
|------|------|
| 학습 대상 | 전체 파라미터 (~1.5B) |
| 정밀도 | float16 |
| gradient_checkpointing | True (메모리 절약) |

> 💡 CPT는 모델의 기본 지식을 변경하는 작업이므로 FFT가 정석입니다.  
> LoRA는 어댑터만 학습하므로 새 지식이 깊이 반영되기 어렵습니다.

````python
# FFT 학습 파라미터 확인
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"✅ FFT 설정 확인")
print(f"📌 전체 파라미터: {total_params:,}")
print(f"📌 학습 가능 파라미터: {trainable_params:,}")
print(f"📌 학습 비율: {trainable_params/total_params*100:.1f}% (FFT = 100%)")
print_gpu_memory("FFT 준비 완료")
````

````python
# FFT는 별도 어댑터 적용이 필요 없습니다
# 모델 전체 파라미터가 그대로 학습 대상입니다
print("✅ FFT 모드: 모델 전체 파라미터를 학습합니다")
````

## 5️⃣ 학습 실행 (Trainer API)

RTX 4060에 안전한 FFT 학습 설정:

- 🔹 `per_device_train_batch_size=1` - 메모리 절약
- 🔹 `gradient_accumulation_steps=8` - 실효 배치 크기 = 8
- 🔹 `gradient_checkpointing=True` - 메모리 절약 (속도 약간 희생)
- 🔹 `optim="adamw_bnb_8bit"` - 8bit 옵티마이저로 메모리 ~2.8GB 절약
- 🔹 `fp16=False` - 혼합 정밀도 학습
- 🔹 `learning_rate=5e-6` - **매우 낮은 학습률** (파국적 망각 방지)
- 🔹 `num_train_epochs=1` - **1 에폭만** (과적합 방지)

### ⚠️ 파국적 망각(Catastrophic Forgetting)이란?

FFT로 CPT를 할 때, 학습률이 높거나 에폭이 많으면 모델이 **기존 지식을 모두 잊어버리는** 현상입니다.
작은 도메인 데이터에 과적합되면서 한국어/영어 생성 능력 자체가 망가질 수 있습니다.

**방지법:**
1. 학습률을 매우 낮게 (5e-6 이하)
2. 에폭을 1로 제한
3. 충분한 양의 도메인 데이터 사용

> 💡 LoRA CPT는 원본 가중치를 동결하므로 파국적 망각이 없습니다.  
> FFT CPT는 성능이 더 좋지만, 이런 주의가 필요합니다.

### 💡 8bit 옵티마이저가 필요한 이유
```
1.5B FFT 메모리 추정:
  모델 (float16):     ~2.8GB
  옵티마이저 (32bit):  ~5.6GB  ← 이게 문제!
  그래디언트:          ~1.1GB (gradient_checkpointing 적용)
  합계:               ~9.5GB  → RTX 4060 (8GB) OOM!

8bit 옵티마이저 적용 시:
  모델 (float16):     ~2.8GB
  옵티마이저 (8bit):   ~2.8GB  ← 절반으로 줄어듦
  그래디언트:          ~1.1GB
  합계:               ~6.7GB  → RTX 4060 (8GB) OK! ✅
```

> ⚠️ OOM(메모리 부족) 발생 시 `MAX_SEQ_LENGTH`를 512로 줄여보세요.

````python
# 학습 전 모델 응답 저장 (비교용)
print("="*60)
print("📋 학습 전 모델 응답 (Before Training)")
print("="*60)

test_prompts = [
    "LoRA(Low-Rank Adaptation)는 대규모 언어 모델을 효율적으로 파인튜닝하기 위한 기법으로,",
    "트랜스포머(Transformer) 아키텍처의 핵심은 셀프 어텐션 메커니즘으로,",
    "QLoRA는 4bit 양자화와 LoRA를 결합한 방법으로,",
    "RAG(Retrieval-Augmented Generation)는 외부 지식 베이스를 활용하여",
    "Continuous Pretraining은 사전학습된 모델에 도메인 데이터를 추가로 학습시키는 방법으로,"
]

# 생성 파라미터
gen_config = dict(
    max_new_tokens=150,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.2,  # 반복 방지
)

model.eval()
before_responses = []

for prompt in test_prompts:
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, **gen_config)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    before_responses.append(response)
    print(f"\n🔹 프롬프트: {prompt[:50]}...")
    print(f"   응답: {response[:250]}")

model.train()
print("\n" + "="*60)
````

````python
# 학습 인자 설정
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    num_train_epochs=1,                    # 1 에폭 (과적합/파국적 망각 방지)
    per_device_train_batch_size=1,         # RTX 4060 안전 설정
    gradient_accumulation_steps=8,          # 실효 배치 크기 = 8
    learning_rate=5e-6,                     # FFT CPT는 매우 작은 lr 사용 (파국적 망각 방지)
    optim="adamw_bnb_8bit",                # 8bit 옵티마이저 (메모리 ~2.8GB 절약)
    fp16=False,                              # 혼합 정밀도
    logging_steps=1,
    save_strategy="epoch",
    warmup_ratio=0.1,
    lr_scheduler_type="cosine",
    report_to="none",
    remove_unused_columns=False,
    gradient_checkpointing=True,
)

print("✅ TrainingArguments 설정 완료")
print(f"📌 batch_size: {training_args.per_device_train_batch_size}")
print(f"📌 gradient_accumulation: {training_args.gradient_accumulation_steps}")
print(f"📌 effective batch size: {training_args.per_device_train_batch_size * training_args.gradient_accumulation_steps}")
print(f"📌 epochs: {training_args.num_train_epochs}")
print(f"📌 learning_rate: {training_args.learning_rate}")
print(f"📌 optimizer: {training_args.optim} (8bit → 메모리 절약)")
print(f"📌 bf16: {training_args.bf16}")
print(f"\n💡 CPT FFT 핵심: 낮은 학습률(5e-6) + 적은 에폭(1) → 파국적 망각 방지")
````

````python
# Data Collator (Causal LM용 - MLM=False)
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # Causal LM이므로 MLM은 False
)

# Trainer 초기화
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    data_collator=data_collator
)

print("✅ Trainer 초기화 완료")
print_gpu_memory("학습 시작 전")
````

````python
# 학습 실행
print("🚀 Continuous Pretraining 학습 시작!")
print("="*60)

train_result = trainer.train()

print("="*60)
print("✅ 학습 완료!")
print(f"📌 Total steps: {train_result.global_step}")
print(f"📌 Training loss: {train_result.training_loss:.4f}")
print_gpu_memory("학습 완료 후")
````

## 6️⃣ 학습 전후 비교 (도메인 지식 질문)

동일한 프롬프트로 학습 전후의 응답을 비교합니다.

- 🔹 도메인 용어에 대한 이해도 변화 확인
- 🔹 텍스트 생성 품질 비교
- 🔹 도메인 지식이 잘 주입되었는지 평가

````python
# 학습 후 모델 응답 생성
print("="*60)
print("📋 학습 전후 비교 (Before vs After)")
print("="*60)

model.eval()
after_responses = []

for i, prompt in enumerate(test_prompts):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, **gen_config)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    after_responses.append(response)
    
    print(f"\n{'='*60}")
    print(f"🔹 프롬프트: {prompt[:50]}...")
    print(f"\n  [Before] {before_responses[i][:250]}")
    print(f"\n  [After]  {response[:250]}")

print(f"\n{'='*60}")
print("📌 도메인 텍스트로 학습한 후, 관련 용어에 대한 연속 생성이 더 자연스러워졌는지 확인하세요.")
````

## 7️⃣ 모델 저장

FFT로 학습한 전체 모델을 저장합니다.

````python
# 모델 전체 저장
save_path = os.path.join(OUTPUT_DIR, "fft_model")
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)

print(f"✅ 모델 저장 완료: {save_path}")

# 저장된 파일 크기 확인
total_size = 0
for root, dirs, files in os.walk(save_path):
    for f in files:
        fp = os.path.join(root, f)
        size = os.path.getsize(fp)
        total_size += size
        print(f"  📄 {f}: {size/1024/1024:.1f} MB")

print(f"\n📌 전체 모델 크기: {total_size/1024/1024:.1f} MB")
print(f"📌 FFT는 전체 모델을 저장하므로 LoRA 어댑터보다 크기가 큽니다")
````

````python
# GPU 메모리 정리
del model, trainer
gc.collect()
torch.cuda.empty_cache()
print_gpu_memory("메모리 정리 후")
print("✅ GPU 메모리 정리 완료")
````

## 📝 정리 및 핵심 요약

### 이번 실습에서 배운 내용

| 항목 | 내용 |
|------|------|
| Continuous Pretraining | 도메인 텍스트를 Next Token Prediction으로 추가 학습 |
| 데이터 준비 | 텍스트 → 토큰화 → 고정 길이 청크 분할 |
| FFT 학습 | float16 + gradient_checkpointing으로 8GB GPU에서 학습 |
| 학습 결과 | 도메인 용어와 개념에 대한 생성 품질 향상 |

### 핵심 포인트

- 🎯 **Continuous Pretraining은 비구조화 텍스트**를 사용합니다 (instruction 형식 아님)
- 🎯 **Next Token Prediction** 방식: labels = input_ids
- 🎯 **FFT가 CPT의 정석** — 모델의 기본 지식을 바꾸려면 전체 파라미터를 업데이트해야 합니다
- 🎯 **도메인 지식 주입**에 효과적 (의료, 법률, 금융 등)
- 🎯 LoRA CPT도 가능하지만, 지식 반영 깊이가 FFT보다 얕습니다
- 🎯 실제 활용 시에는 **더 많은 도메인 데이터**가 필요합니다

### CPT: FFT vs LoRA

| | FFT (이번 실습) | LoRA |
|---|---|---|
| 지식 반영 | 깊음 (전체 파라미터 업데이트) | 얕음 (어댑터만 업데이트) |
| 메모리 | 1.5B까지 8GB에서 가능 | 7B까지 8GB에서 가능 |
| 저장 크기 | ~3GB (전체 모델) | ~50MB (어댑터만) |
| 적합한 경우 | 도메인 지식 깊이 주입 | 리소스 제한 시 타협안 |
